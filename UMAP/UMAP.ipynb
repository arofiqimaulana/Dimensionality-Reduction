```python
import umap
import numpy as np
import pandas as pd
from sklearn.datasets import load_iris
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt
```

## Load dataset Iris

```python
iris = load_iris()
X = iris.data  # Fitur data (4 dimensi)
y = iris.target  # Label kelas
```

```python
# Standarisasi data
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
# Terapkan UMAP untuk mengurangi dimensi menjadi 2
umap_model = umap.UMAP(n_components=2, random_state=42)
X_umap = umap_model.fit_transform(X_scaled)
```

```python
# Visualisasi hasil reduksi dimensi
plt.figure(figsize=(8, 6))
scatter = plt.scatter(X_umap[:, 0], X_umap[:, 1], c=y, cmap='viridis')
plt.colorbar(scatter)
plt.title("Reduksi Dimensi dengan UMAP (Iris Dataset)")
plt.xlabel('Dimensi 1')
plt.ylabel('Dimensi 2')
plt.show()
```
